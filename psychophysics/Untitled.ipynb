```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("KBY_18-Nov-2022 00:37:32_session_data.csv")
df.head()
```

```python
data = {}
for c in df['Trial']:
    temp = df[df.Trial == c].values[0]
    data[temp[0]] = temp[1:]
data
```

```python
df = pd.DataFrame.from_dict(data, orient='index').T
df.head()
```

```python
tmp = df.iloc[:, :5]
sns.lineplot(x = tmp.index, y = tmp['Gray-0'])
sns.lineplot(x = tmp.index, y = tmp['Gray-40'])
sns.lineplot(x = tmp.index, y = tmp['Gray-80'])
sns.lineplot(x = tmp.index, y = tmp['Gray-120'])
sns.lineplot(x = tmp.index, y = tmp['Gray-160'])
plt.legend()
```
