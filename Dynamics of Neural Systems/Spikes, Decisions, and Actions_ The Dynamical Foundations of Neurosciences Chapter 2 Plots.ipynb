```python
import numpy as np
import matplotlib.pyplot as plt
```

# Q2

```python
def r(P,t):
    return 5*np.exp(-t/20)+(50*P**4)/(15**4+P**4)*(1-np.exp(-t/20))
```

```python
t = np.arange(0,100,0.5)

p10 = [r(10, i) for i in t]
p20 = [r(20, i) for i in t]
p30 = [r(30, i) for i in t]
```

```python
plt.plot(t, p10, 'k', label = 'P = 10')
plt.plot(t, p20, '--k', label = 'P = 20')
plt.plot(t, p30, '-.k', label = 'P = 30')
plt.xlabel('t')
plt.ylabel('R(t)')
plt.legend()
```

# Q4

```python
def V(t):
    return -70*np.exp(-22.4*t) + 35*(1-np.exp(-22.4*t))

def V2(t):
    return 35*np.exp(-8*t) + -90*(1-np.exp(-8*t))

def V3(t):
    return -90*np.exp(-2.8*t) + -70*(1-np.exp(-2.8*t))

t1 = np.arange(0,1,0.01)
v1 = [V(i) for i in t1]

t2 = np.arange(0,3,0.01)
v2 = [V2(i) for i in t2]

t3 = np.arange(0,4,0.01)
v3 = [V3(i) for i in t3]

plt.plot(t1, v1, '--k')
plt.plot(t2+1, v2, '--k')
plt.plot(t3+4, v3, 'k--')
plt.xlabel('t')
plt.ylabel('V(t)')
```

# Q5

```python
def V(t, N):
    return -75*np.exp(-(2*N+1)*t/12.5) - 75/(2*N+1)*(1-np.exp(-(2*N+1)*t/12.5))

t = np.arange(0,1,0.01)
N = np.array(range(1, 13))

Vs = [V(1, n) for n in N]

plt.scatter(N, Vs)
plt.plot(N, Vs, 'k--')
plt.grid()
plt.legend()
plt.xlabel('N')
plt.ylabel('V(t)')
```
